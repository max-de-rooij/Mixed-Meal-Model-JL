```julia
using DataFrames, StatsPlots, DelimitedFiles, Statistics
```

```julia
sim_julia = readdlm("Results/Windows_ShitMachine_Benchmark_Simulation.txt")./1e9 # Julia runs in nanoseconds
sim_matlab = readdlm("Results/Windows_ShitMachine_Benchmark_Simulation_Matlab.txt", ',')'

sim = boxplot(sim_julia, yaxis=:log, label="Julia", color=RGB(20/255,55/255,1), bar_width=0.3)
boxplot!(repeat([1.4], 10000), sim_matlab, color = RGB(250/255, 86/255, 55/255), bar_width=0.3, label="MATLAB", size=(200,200), xaxis=false, grid=false, title="Simulation", titlefontfamily="Helvetica Bold",ylabel="Time (s)", titlefontsize=12, guidefontsize=9, guidefontfamily="Helvetica Bold")
savefig("Figures/Simulation_Benchmark.svg")
#violin(sim_julia)
#median(sim_matlab)/median(sim_julia)
```

```julia
paramest_julia = readdlm("Results/Windows_ShitMachine_Benchmark_ParamEstim.txt")./1e9
paramest_matlab = readdlm("Results/Windows_ShitMachine_Benchmark_ParamEstim_Matlab.txt", ',')'

paramest = boxplot(paramest_julia, yaxis=:log, label="", color=RGB(20/255,55/255,1), bar_width=0.3)
boxplot!(repeat([1.4], 100), paramest_matlab,  color = RGB(250/255, 86/255, 55/255),label="", size=(200,200), bar_width=0.3, xaxis=false, grid=false, title="Parameter Estimation", titlefontfamily="Helvetica Bold",ylabel="Time (s)", titlefontsize=12, guidefontsize=9, guidefontfamily="Helvetica Bold")
savefig("Figures/ParamEst_Benchmark.svg")

median(paramest_matlab)/median(paramest_julia)
```

```julia
pla_julia = readdlm("Results/Windows_ShitMachine_Benchmark_PLA.txt")./1e9
pla_matlab = readdlm("Results/Windows_ShitMachine_Benchmark_PLA_Matlab.txt", ',')'

pla = boxplot(pla_julia, yaxis=:log, label="", color=RGB(20/255,55/255,1), bar_width=0.3)
boxplot!(repeat([1.4], 10), pla_matlab,  color = RGB(250/255, 86/255, 55/255),label="", bar_width=0.3, size=(200,200), xaxis=false, grid=false, title="Identifiability Analysis", titlefontfamily="Helvetica Bold",ylabel="Time (s)", titlefontsize=12, guidefontsize=9, guidefontfamily="Helvetica Bold")

savefig("Figures/PLA_Benchmark.svg")

median(pla_matlab)/median(pla_julia)
```

```julia
sens_julia = readdlm("Results/Windows_ShitMachine_Benchmark_Sensitivity.txt")./1e9
sens_matlab = readdlm("Results/Windows_ShitMachine_Benchmark_Sensitivity_Matlab.txt", ',')'

sens = boxplot(sens_julia, yaxis=:log, label="", color=RGB(20/255,55/255,1), bar_width=0.3)
boxplot!(repeat([1.4], 100), sens_matlab, color = RGB(250/255, 86/255, 55/255), label="", bar_width=0.3, size=(200,200), xaxis=false, grid=false, title="Sensitivity Analysis", titlefontfamily="Helvetica Bold",ylabel="Time (s)", titlefontsize=12, guidefontsize=9, guidefontfamily="Helvetica Bold")

savefig("Figures/Sens_Benchmark.svg")

median(sens_matlab)/median(sens_julia)
```
